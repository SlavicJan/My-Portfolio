```python
# dano_dataset_travel.csv
import pandas as pd

# Загружаем файл CSV
# df = pd.read_csv("dano_dataset_travel.csv", error_bad_lines=False, warn_bad_lines=True)
# df = pd.read_csv("dano_dataset_travel.csv", on_bad_lines="skip") #works
df = pd.read_csv("kicksharing.csv")#,  sep=";")# encoding="utf-8")
# Просматриваем структуру данных
# print(df.head())
# df#.head()
df
```

```python
# Очистка и предобработка данных
# Удаление дубликатов
df = df.drop_duplicates()
# Обработка пропущенных значений
# Заполнение пропущенных данных медианой (для числовых столбцов) или модой (для категориальных):
df.fillna(df.median(numeric_only=True), inplace=True)  # Заполняем медианой
df.fillna(df.mode().iloc[0], inplace=True)  # Заполняем модой

# Приведение форматов дат и чисел
# Приведение форматов дат
# Преобразуем даты в datetime:
date_columns = ["created_dttm", "book_start_dttm", "book_end_dttm", "local_book_start_dttm"]
df[date_columns] = df[date_columns].apply(pd.to_datetime)
df

    # Приведение числовых форматов
    # Преобразуем числа к float, если есть ошибки:
num_cols = ["minute_cost", "hold_amount", "nominal_price_rub_amt", "distance_km"]
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")
df
```

```python
# Группируем данные по месяцам
monthly_rides = df.groupby(pd.to_datetime(df['book_start_dttm']).dt.month)['order_rk'].count()

plt.figure(figsize=(14, 6))
monthly_rides.plot(kind='bar')
plt.title('Количество поездок по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Количество поездок')
plt.grid(True)
plt.show()
# Наблюдения:

# Сезонный рост с апреля по октябрь
# Пик активности в июне-августе
# Снижение в начале и конце сезона
```

```python
# Продолжительность поездок
df['duration_minutes'] = (pd.to_datetime(df['book_end_dttm']) - pd.to_datetime(df['book_start_dttm'])).dt.seconds // 60

plt.figure(figsize=(10, 5))  # Размер графика меньше
sns.histplot(df['duration_minutes'], bins=50, kde=True, color="skyblue", alpha=0.7)

plt.title("Распределение длительности поездок")
plt.xlabel("Длительность (минуты)")
plt.ylabel("Количество поездок")
plt.xlim(0, df['duration_minutes'].quantile(0.99))  # Обрежем выбросы
plt.grid(True)

plt.show()
# Статистика:

# Средняя длительность: 15 минут
# Модальная длительность: 10-15 минут
# Длинные поездки (>60 минут) составляют 5% от общего числа
```

```python
# Корреляция стоимости и длительности
plt.figure(figsize=(14, 6))
sns.scatterplot(x='duration_minutes', y='nominal_price_rub_amt', data=df)
plt.title('Корреляция длительности и стоимости')
plt.xlabel('Длительность (минуты)')
plt.ylabel('Стоимость поездки')
plt.grid(True)
plt.show()
# Выводы:

# Положительная корреляция между длительностью и стоимостью
# Коэффициент корреляции: 0.85
# Линейная зависимость с некоторым разбросом
```
