### Общий прогресс студента

Всего заданий на курсе - 180

```python
data[:2]
```

```python
scores = {} # Создаем пустой словарь для хранения баллов студентов

for row in data[1:]: # Проходимся по каждой строке данных, кроме заголовка
  user = row[1] # Извлекаем user_id
  problem = row[2] # Извлекаем problem_id
  is_correct = 1 - eval(row[3]) # Проверяем, было ли решение правильным (0 - правильно, 1 - неправильно)

  if not is_correct or row[-2] == 'Simple': # Пропускаем неправильные решения и задачи типа 'Simple'
    continue

  if user not in scores: # Если пользователя еще нет в словаре scores
    scores[user] = set([problem]) # Добавляем пользователя и создаем множество с решенной задачей
  else: # Если пользователь уже есть в словаре scores
    scores[user].add(problem) # Добавляем решенную задачу в множество пользователя
```

```python
scores
```

```python
for k, v in scores.items(): # Проходимся по каждому элементу в словаре scores (где k - user_id, v - множество решенных задач)
  scores[k] = f'{round(len(v) / 180 * 100, 1)}%' # Вычисляем процент решенных задач (количество решенных задач / общее количество задач * 100), округляем до одного знака после запятой и форматируем как строку с символом '%'
```

```python
scores
```

```python
dict(sorted(scores.items(), key=lambda x: eval(x[1][:-1]), reverse=True)) # Сортируем словарь scores по значению (проценту решенных задач), предварительно удаляя символ '%' и преобразуя строку в число для корректной сортировки. Результат преобразуется обратно в словарь.
```

### Табель с успеваемостью студентов

```python
data[:2]
```

```python
problems = set() # Создаем пустое множество для хранения уникальных ID задач

for row in data[1:]: # Проходимся по каждой строке данных, кроме заголовка
  if row[-2] == 'Homework': # Проверяем, является ли задача домашним заданием (Homework)
    problems.add(row[2]) # Добавляем ID задачи в множество problems, если это домашнее задание
```

```python
len(problems)
```

```python
marks = {} # Создаем пустой словарь для хранения информации о решенных задачах каждого студента

for row in data[1:]: # Проходимся по каждой строке данных, кроме заголовка
  user = row[1] # Извлекаем user_id
  problem = row[2] # Извлекаем problem_id
  is_correct = 1 - eval(row[3]) # Проверяем, было ли решение правильным

  if not is_correct or row[-2] != 'Homework': # Пропускаем неправильные решения и задачи, которые не являются домашними заданиями
    continue

  if user in marks: # Если пользователь уже есть в словаре marks
    marks[user].add(problem) # Добавляем решенную задачу в множество пользователя
  else: # Если пользователя еще нет в словаре marks
    marks[user] = set(problem) # Добавляем пользователя и создаем множество с решенной задачей
```

```python
marks
```

```python
users = set() # Создаем пустое множество для хранения уникальных ID пользователей

for row in data[1:]: # Проходимся по каждой строке данных, кроме заголовка
  users.add(row[1]) # Добавляем user_id в множество users
```

```python
problems = list(problems)
```

```python
marks_for_print = {} # Создаем пустой словарь для форматированного вывода оценок

for user in users: # Проходимся по каждому уникальному пользователю
  for problem in problems: # Проходимся по каждой уникальной задаче
    is_solved = 'v' if problem in marks.get(user, []) else 'x' # Проверяем, решена ли задача пользователем. Если пользователь есть в словаре marks и задача есть в его множестве решенных задач, ставим 'v', иначе 'x'. Используем marks.get(user, []) для безопасного доступа к ключу user (если пользователя нет, вернется пустой список)
    if not user in marks_for_print: # Если пользователя еще нет в словаре marks_for_print
      marks_for_print[user] = [is_solved] # Добавляем пользователя и создаем список с результатом по текущей задаче
    else: # Если пользователь уже есть в словаре marks_for_print
      marks_for_print[user].append(is_solved) # Добавляем результат по текущей задаче в список пользователя
```

```python
marks_for_print
```

```python
max([len(x) for x in users]) # Находим максимальную длину user_id
```

```python
max([len(x) for x in problems]) # Находим максимальную длину строкового представления каждого problem_id в списке problems и возвращаем наибольшее значение. Это нужно для форматирования таблицы успеваемости, чтобы выровнять столбцы.
```

```python
'user_id  ' + "| ".join([id + " "*(5 - len(id)) for id in problems]) # Формируем строку заголовка для таблицы успеваемости, добавляя user_id и отформатированные ID задач с пробелами для выравнивания
```

```python
res = [] # Создаем пустой список с именем 'res'. Этот список будет использоваться для хранения каждой строки будущей таблицы успеваемости студентов.

# Формируем строки для каждого пользователя в таблице
for k, v in marks_for_print.items(): # Проходимся по каждому элементу в словаре 'marks_for_print'. Словарь содержит user_id в качестве ключей (обозначено как 'k') и список результатов ('v' - 'v' для решенной задачи, 'x' для нерешенной) для каждой задачи в качестве значений.
  s = k + ' '*(9 - len(k)) # Формируем начальную часть текущей строки таблицы. Берем user_id (k), вычисляем количество пробелов, необходимое для того, чтобы user_id занял ровно 9 символов (это нужно для выравнивания первого столбца в таблице), и добавляем эти пробелы к user_id.
  s += "   |   ".join(v) # Добавляем к строке 's' результаты по задачам для текущего пользователя. Элементы списка 'v' (отметки 'v' или 'x') объединяются в одну строку, разделенную строкой "   |   ". Это формирует столбцы с отметками для каждой задачи после user_id.
  res.append(s) # Добавляем полностью сформированную строку для текущего пользователя в список 'res'.

# Формируем строку заголовка для таблицы с ID задач
# Сначала создаем список отформатированных ID задач
header_problems = [id + " "*(6 - len(id)) for id in problems] # Создаем новый список. Для каждого ID задачи ('id') в списке 'problems' вычисляем количество пробелов, необходимое для того, чтобы этот ID занял 6 символов (учитывая максимальную длину ID задачи и добавляя пробелы для выравнивания). Затем объединяем каждый ID с соответствующим количеством пробелов.
# Объединяем отформатированные ID задач в одну строку заголовка, добавляя разделители
header_row = 'user_id' + "| ".join(header_problems) # Создаем строку заголовка. Начинаем со строки 'user_id', затем объединяем все отформатированные ID задач из списка 'header_problems', используя строку "| " в качестве разделителя.
# Вставляем строку заголовка в начало списка 'res'
res.insert(0, header_row) # Вставляем сформированную строку заголовка ('header_row') в начало списка 'res' (позиция 0). Таким образом, строка заголовка станет первой строкой при последующем выводе таблицы.

# старый код:
# res = []

# for k, v in marks_for_print.items():
#   s = k + ' '*(9 - len(k))
#   s += "   |   ".join(v)
#   res.append(s)

# res.insert(0, 'user_id' + "| ".join([id + " "*(6 - len(id)) for id in problems]))
```
