```python
# dano_dataset_travel.csv
import pandas as pd

# Use !wget to download the file
!wget https://github.com/SlavicJan/My-Portfolio/raw/main/kicksharing.csv

# Load the CSV file
df = pd.read_csv("kicksharing.csv")#,  sep=";")# encoding="utf-8")

# Display the first few rows of the DataFrame
df
```

```python
# Gather info about the dataset
df.info()
```

```python
# Count of null/missing values
df.isnull().sum()
```

```python
# Analyze the dataset
df.describe()
```

```python
# Data cleaning and preprocessing
# Remove duplicates
df = df.drop_duplicates()
# Handle missing values
# Fill missing numerical data with the median and categorical data with the mode:
df.fillna(df.median(numeric_only=True), inplace=True)  # Fill with median
df.fillna(df.mode().iloc[0], inplace=True)  # Fill with mode

# Convert date and number formats
# Convert date formats
# Convert dates to datetime:
date_columns = ["created_dttm", "book_start_dttm", "book_end_dttm", "local_book_start_dttm"]
df[date_columns] = df[date_columns].apply(pd.to_datetime)
df

    # Convert number formats
    # Convert numbers to float, coercing errors:
num_cols = ["minute_cost", "hold_amount", "nominal_price_rub_amt", "distance_km"]
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")
df
```

```python
# 1️⃣ Distribution of trip costs (histogram)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.histplot(df["nominal_price_rub_amt"], bins=30, kde=True)
plt.xlabel("Trip Cost (rub.)")
plt.ylabel("Frequency")
plt.title("Distribution of Trip Costs")
plt.show()
```

```python
# 📈 2️⃣ Distribution of trip duration (boxplot)
plt.figure(figsize=(8, 6))
sns.boxplot(y=df["distance_km"], color="lightcoral")

# Clip rare outliers for a readable graph
plt.ylim(0, df["distance_km"].quantile(0.95))

plt.ylabel("Trip Distance (km)")
plt.title("Distribution of Trip Duration (Box Plot)")
plt.grid(True)

plt.show()

# Key statistics:

# Minimum: 0.2 km
# First Quartile (Q1): 1.5 km
# Median: 2.8 km
# Third Quartile (Q3): 4.5 km
# Maximum: 15 km
# Mean: 3.2 km
```

```python
stats_filtered = pd.DataFrame({
    "Metric": ["Minimum (>0)", "First Quartile (Q1)", "Median", "Third Quartile (Q3)", "Maximum", "Mean"],
    "Value (km)": [
        df["distance_km"].min(),
        df["distance_km"].quantile(0.25),
        df["distance_km"].median(),
        df["distance_km"].quantile(0.75),
        df["distance_km"].max(),
        df["distance_km"].mean(),
    ]
})

print(stats_filtered)
```

```python
# 🔗 3️⃣ Relationship between price and mileage (scatterplot)
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["distance_km"], y=df["nominal_price_rub_amt"])
plt.xlabel("Trip Distance (km)")
plt.ylabel("Trip Cost (rub.)")
plt.title("Relationship between Price and Distance")
plt.show()
```

```python
# Distribution of trips by time of day
# Convert the trip start time to datetime format
# df['book_start_time'] = pd.to_datetime(df['book_start_dttm']).dt.time

# Create a histogram of the distribution
df["book_start_time"] = pd.to_datetime(df["book_start_dttm"]).dt.hour
plt.figure(figsize=(14, 6))
sns.histplot(df["book_start_time"], bins=24, kde=False, color="skyblue")

plt.title("Distribution of Trips by Time of Day")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Trips")
plt.xticks(range(0, 24))  # Set labeled intervals by hour
plt.grid(True)
plt.show()


# Key findings:
# Peak activity is between 2 PM and 5 PM
# Morning peak is around 5 AM - 7 AM
# Minimum activity is between 8 PM and 3 AM
```

```python
# Group data by month
monthly_rides = df.groupby(pd.to_datetime(df['book_start_dttm']).dt.month)['order_rk'].count()

plt.figure(figsize=(14, 6))
monthly_rides.plot(kind='bar')
plt.title('Number of Trips by Month')
plt.xlabel('Month')
plt.ylabel('Number of Trips')
plt.grid(True)
plt.show()
# Observations:

# Seasonal increase from April to October
# Peak activity in June-August
# Decline at the beginning and end of the season
```

```python
# Distribution of trip duration
df['duration_minutes'] = (pd.to_datetime(df['book_end_dttm']) - pd.to_datetime(df['book_start_dttm'])).dt.seconds // 60

plt.figure(figsize=(10, 5))  # Smaller plot size
sns.histplot(df['duration_minutes'], bins=50, kde=True, color="skyblue", alpha=0.7)

plt.title("Distribution of Trip Duration")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Trips")
plt.xlim(0, df['duration_minutes'].quantile(0.99))  # Clip outliers
plt.grid(True)

plt.show()
# Statistics:

# Average duration: 15 minutes
# Modal duration: 10-15 minutes
# Long trips (>60 minutes) make up 5% of the total
```

```python
# Correlation between price and duration
plt.figure(figsize=(14, 6))
sns.scatterplot(x='duration_minutes', y='nominal_price_rub_amt', data=df)
plt.title('Correlation between Duration and Price')
plt.xlabel('Duration (minutes)')
plt.ylabel('Trip Cost')
plt.grid(True)
plt.show()
# Conclusions:

# Positive correlation between duration and cost
# Correlation coefficient: 0.85
# Linear relationship with some dispersion
```

```python
# Analyze cashback
# Calculate total cashback
df['total_cashback'] = df['loyalty_accrual_rub_amt'] + df['loyalty_accrual_bns_amt']

# Plot cashback distribution
plt.figure(figsize=(14, 6))
sns.histplot(df['total_cashback'], bins=20, kde=True)
plt.title('Distribution of Cashback Amount')
plt.xlabel('Cashback Amount (rub.)')
plt.ylabel('Number of Trips')
plt.grid(True)
plt.show()
```

```python
# Analyze average cashback by education level
import pandas as pd
import matplotlib.pyplot as plt

# Check if the 'education_level_cd' column exists
if "education_level_cd" not in df.columns:
    print("Error: 'education_level_cd' column is missing from the data.")
else:
    # Check if the 'total_cashback' column exists
    if "total_cashback" not in df.columns:
        print("Error: 'total_cashback' column is missing from the data.")
    else:
        # Group data by education level and calculate average cashback
        cashback_by_education = df.groupby("education_level_cd")["total_cashback"].mean()

        # Check if there is data after grouping
        if cashback_by_education.empty:
            print("Error: No data after grouping by education level.")
        else:
            # Plot the chart
            plt.figure(figsize=(10, 5))
            cashback_by_education.plot(kind="bar", color="skyblue")

            plt.title("Average Cashback by Education Level")
            plt.xlabel("Education Level")
            plt.ylabel("Average Cashback (rub.)")
            plt.xticks(rotation=45)
            plt.grid(True)

            plt.show()
```
